## How to Make a Box Plot in ggplot2

This tutorial creates a unique dataset using the ages of 20 study subjects from two groups (A,B) to make a box plot. 


### Create the data set

Create a vector containing the group assignments for 20 participants.

```r
v1 <- c(rep("A",10),rep("B",10))
v1
```

Create a second vector containing random ages for 20 people between 15 and 65 years old.

```r
v2 <- c(runif(20,min=15, max = 65))
v2
```

The numbers generated above have decimals, we just want ages as integers so round all the values of v2.

```r
v2 <- round(v2)
v2
```

We have a vector of group assignments and a vector of ages. Now combine them to make a data frame.

```r
df <- data.frame(v1,v2)
df
```

Now name the columns for clarity and ease when making plots. 

```r
colnames(df) <- c("Group","Age")
df
```

### Make a box plot

Load the package "ggplot2" to make the plot.

```r
library(ggplot2)
```

Start with a basic box plot where Group is on the x axis and Age is on the Y axis, assign this plot to an object we can call later to add other features.
In the basic plot we can color our boxes based on which group it belongs to. 
The ggplot() function creates the basic structure of the plot and the geom_boxplot() function indicates how we'd like to display the data (as a boxplot). 

```r
Plot <- ggplot(df, aes(x=Group, y = Age, color = Group)) + geom_boxplot()
Plot
```

Now that we have our basic plot, we can add stat_summary() to display statistical values such as the mean ages for both groups. We want the mean to be displayed by a point on the map in a diamond shape (shape=23) and we want it a bit bigger than default (size = 5).

```r
Plot + stat_summary(fun=mean, geom="point", shape=23, size=5)
```

Finally, we are curious where all our participants fall in these box plots so we can overlap points for each individual. We will use geom_jitter() to avoid overlapping points. We can again customize what shape and size the points will be and how much "jitter" to use.

Hint: If you want the dots more spread out try position=position_jitter(0.2)

```r
Plot + stat_summary(fun=mean, geom="point", shape=23, size=5) + geom_jitter(shape=16, size= 3, position=position_jitter(0.1))
```
